```python
from pyspark.sql.types import StructType,StructField, StringType,IntegerType,DoubleType
```

```python
EmpSchema = StructType([  
    StructField('Empno', IntegerType(), True), 
    StructField('Empname', StringType(), True),    
     StructField('salary', DoubleType(), True) 
])
```

```python
employeedf=spark.read.option("header","true").schema(EmpSchema).csv("/home/labuser/Documents/datasets/employee.csv")
```

```python
employeedf.show()
```

```python
employeedf.printSchema()
```

```python
employeedf_malformed=spark.read.option("header","true").option("mode","dropmalformed").schema(EmpSchema).csv("/home/labuser/Documents/datasets/employee.csv")
```

```python
employeedf_malformed.show()
```

```python
employeedf_failfast=spark.read.option("header","true").option("mode","failfast").schema(EmpSchema).csv("/home/labuser/Documents/datasets/employee.csv")
```
