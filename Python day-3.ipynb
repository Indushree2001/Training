```python
import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import interact, widgets
```

```python
def dynamic_scatter_plot(x_col, y_col):
    plt.figure(figsize=(10, 6))
    plt.scatter(moviedf[x_col], moviedf[y_col], c=moviedf['rating'], cmap='viridis', alpha=0.6)
    plt.colorbar(label='Rating')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title(f'Scatter Plot of Movies ({x_col} vs. {y_col})')
    plt.show()
```

```python
x_column_widget = widgets.Dropdown(options=moviedf.columns, description='X-axis:')
y_column_widget = widgets.Dropdown(options=moviedf.columns, description='Y-axis:')

# Create an interactive plot using ipywidgets
interact(dynamic_scatter_plot, x_col=x_column_widget, y_col=y_column_widget)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans

# Load a sample dataset from scikit-learn
data = datasets.load_iris()
X = data.data  # Features
y = data.target  # Target labels

# Fit a KMeans clustering model
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
labels = kmeans.labels_

# Create a DataFrame with the data and labels
df = pd.DataFrame({'Feature1': X[:, 0], 'Feature2': X[:, 1], 'Cluster': labels})

# Create scatter plots for each cluster
plt.figure(figsize=(10, 6))
colors = ['r', 'g', 'b']
for cluster_num, color in zip(range(3), colors):
    cluster_data = df[df['Cluster'] == cluster_num]
    plt.scatter(cluster_data['Feature1'], cluster_data['Feature2'], c=color, label=f'Cluster {cluster_num}')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering of Iris Data')
plt.legend()
plt.show()

```
